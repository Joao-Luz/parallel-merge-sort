```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('results.csv')
```

```python
mean = df.groupby(['n', 'k'], as_index=False).mean()
```

```python
fig, ax = plt.subplots(ncols=3, figsize=(15, 4))

fig.supxlabel('Threads')
fig.supylabel('Time (ms)')

for i,n in enumerate(mean['n'].unique()):
    mean[mean['n'] == n].plot(x='k', y=['sort', 'merge', 'total'], kind='bar', ax=ax[i]);

```
